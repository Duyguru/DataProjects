```python
numeric_features = [ feature for feature in df_clean.columns if df_clean[feature].dtype != 'object'] 
categoric_features = [ feature for feature in df_clean.columns if df_clean[feature].dtype == 'object']
```

```python
plt.figure(figsize=(15,10))

for i in range(0, len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=df_clean[numeric_features[i]],color = "b", fill=True)
    plt.xlabel(numeric_features[i])
    plt.tight_layout()

plt.show()

```

```python
plt.figure(figsize=(15,5))

category = ["Type", "Content Rating"]

for i in range(0, len(category)):
    plt.subplot(1,2,i+1)
    sns.countplot(x=df_clean[category[i]],color = "b", fill=True)
    plt.xlabel(category[i])
    plt.tight_layout()

plt.show()
```
